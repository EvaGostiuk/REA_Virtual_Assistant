```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
import torch
from transformers import pipeline

model = "meta-llama/Llama-2-7b-chat-hf" # meta-llama/Llama-2-7b-chat-hf
tokenizer = AutoTokenizer.from_pretrained(model, use_auth_token=True)

llama_pipeline = pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

```python
def get_response(prompt: str):
    sequences = llama_pipeline(
        prompt,
        do_sample=True,
        top_k=10,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_length=1024,
        truncation=True
    )
    return sequences[0]['generated_text'][len(prompt):]
```

```python
import re
def extract_value(text):
    match = re.search(r'!#(.*?)!#', text, re.DOTALL)
    if match:
        return match.group(1).strip()
    else:
        return ""
```

```python
# Intent Check
def getPromptId(message: str, intentProposed: str):
    SYSTEM_PROMPT = f"""<s>[INST] <<SYS>> You know that proposed intent is {intentProposed}. Select most appropriate prompt tag for the user message and put it in following form "!# message_tag !#". Do not add words that are not present in the original messages or answer user message directly.
    -> paper_search_by_query
    -> paper_recommend_similar
    -> paper_add_to_collection_tag
    -> paper_list_tags_collections
    -> paper_add_annotation_comment
    -> paper_list_annotations
    -> paper_ask_metadata_paper  {answer questions about currently selected paper}
    -> paper_ask_question_annotation  {ask opinion or definition about topics in currently selected paper }<</SYS>> """

    responce = get_response(SYSTEM_PROMPT + f"{message} [/INST]")
    intentDecided = extract_value(responce)
    return {"prompt": intentDecided}
```

```python
# Extract Values
def extractValue(message: str, intentDataType: str):
    intent_actions = {
    #search
    "paper_search_by_query":
    """<s>[INST] <<SYS>> You are helping to discover academic papers. Extract query from user messages that can be put into search bar. Return extracted queries in a form \"!# {query } !#\" .
    Do not add words that are not present in the original messages or answer user message directly.
    Examples:
    \"I would like to learn more about relational databases.\" => \"Great! query is: !# {relational databases} !#
    \"Can you find me a papers about relational databases?\" => \"#Great! query is: !# {relational databases} !# \"
    \"Search relational databases\" => \"Great! query is: !# {relational databases} !#\"
    \"relational databases\" => \"Great! query is: !# {relational databases} !#\"

    In case user did not provide topic content, strat your message with \"!#  `error_not_found` !#\  and ask about what user want to record in the comment.
    \"I would like to find a paper\" => \"Great! query is: !# { `error_not_found` Which topic are you interested in?} !#\" <</SYS>> """,
    #annotate
    "paper_add_annotation_comment": """ <s>[INST] <<SYS>> User wants to leave comment about the paper. Extract content of the annotation/comment and return extracted value in a form \"!# { annotation } !#\ .
    Do not add words that are not present in the original messages or answer user message directly. Examples:
    \"Can you annotate this paper with comment \"This paper would be interesting for CSI5180 presentation.\" "\" => \"Great! annotation is: !# {This paper would be interesting for CSI5180 presentation.} !#\"
    \"Annotate that this paper will be useful for CSI5180 presentations. \" => \"Great! annotation is: !# {This paper would be interesting for CSI5180 presentation.} !#\"

    In case user did not provide comment content, strat your message with \"!#  `error_not_found` !#\  and ask about what user want to record in the comment.
    \"Annotate this paper \" => \"!# {`error_not_found` What would you like to add to your comment?} !#\"  <</SYS>>
    """,
    #tag
    "paper_add_to_collection_tag": """ <s>[INST] <<SYS>> User wants to save the paper to spefific collection using tags.
    Extract name of the collection/tag and return extracted value in a form \"!# { tag } !#\ .
    In case user did not provide tag name, strat your message with \"!#  `error_not_found` !#\  and request user to provide name of the tag/collection.
    Do not add words that are not present in the original messages or answer user message directly. Examples:
    \"Tag this paper as \"CSI5180_presentation\" \" => \"Great! tag is: !# {CSI5180_presentation} !#\"
    \"Add currently selected paper to CSI5180_presentation collection\" => \"Great! tag is: !# {CSI5180_presentation} !#\"

    \"I want to tag this paper \" => \"!# {`error_not_found` What is the name of the tag?} !#\"  <</SYS>>
    """
    }

    SYSTEM_PROMPT = intent_actions[intentDataType]
    responce = get_response(SYSTEM_PROMPT + f"{message} [/INST]")
    val = extract_value(responce)
    return {"value":val}
```
